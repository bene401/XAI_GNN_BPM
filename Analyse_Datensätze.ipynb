```python
def count_class(ds):
  class_0 = 0
  class_1 = 0
  for instanz in ds:
    if int(instanz.y.item()) == 0:
      class_0 += 1
    else:
      class_1 += 1
  return class_0, class_1
```

```python
print(f"Review ab: Klasse 0: {count_class(review_ab)[0]}, Klasse 1: {count_class(review_ab)[1]}  \n"
      f"Review eb: Klasse 0: {count_class(review_eb)[0]},  Klasse 1: {count_class(review_eb)[1]} \n"
      f"Loan ab: Klasse 0: {count_class(loan_ab)[0]},  Klasse 1: {count_class(loan_ab)[1]} \n"
      f"Loan eb: Klasse 0: {count_class(loan_eb)[0]},  Klasse 1: {count_class(loan_eb)[1]} ")
```

## Durchschnittliche Knoten-Zahl

```python
def count_nodes_absolut(ds):
  nodes_sum = 0
  for instanz in ds:
    nodes = (instanz.x.shape[0])
    nodes_sum += nodes
  return nodes_sum
```

```python
print(f"Review ab: {np.round(count_nodes_absolut(review_ab)/len(review_ab))} \n"
      f"Review eb: {np.round(count_nodes_absolut(review_eb)/len(review_eb))}\n"
      f"Loan ab: {np.round(count_nodes_absolut(loan_ab)/len(loan_ab))}\n"
      f"Loan eb: {np.round(count_nodes_absolut(loan_eb)/len(loan_eb))}")
```

## Maximale Knoten-Anzahl

```python
def get_max_nodes_number(ds):
  max = 0
  for instanz in ds:
    nodes = instanz.x.shape[0]
    if nodes > max:
      max = nodes
  return max
```

```python
print(f"Review ab: {get_max_nodes_number(review_ab)} \n"
      f"Review eb: {get_max_nodes_number(review_eb)}\n"
      f"Loan ab: {get_max_nodes_number(loan_ab)}\n"
      f"Loan eb: {get_max_nodes_number(loan_eb)}")
```

## Minimale Knoten-Anzahl

```python
def get_min_nodes_number(ds):
  min = ds[0].x.shape[0]
  for instanz in ds:
    nodes = instanz.x.shape[0]
    if nodes < min:
      min = nodes
  return min
```
